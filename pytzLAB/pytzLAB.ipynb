## TimeZone Conversion with PYTZ

1. Install pytz `pip install pytz`

```python
import pytz
from datetime import datetime
```

```python
# create a function with pytz to convert timezones
# accepts 3 parameters; 
# dt(date/time), 
# from_tz(for current timezone), 
# to_tz(timezone to convert time to)
def convert_time_zone(dt, from_tz, to_tz):
# create timezone objects
    from_tz = pytz.timezone(from_tz)
    to_tz = pytz.timezone(to_tz)
# Convert datetime object to the from timezone if necessary
# settimg the tz attribute
    dt = dt.replace(tzinfo=from_tz)

# converting datetime to new timzone
    dt = dt.astimezone(to_tz)

    return dt
```

```python
# using created function
today_datetime = datetime.now()

from_timezone = 'US/Eastern'
to_timezone = 'Australia/Melbourne'
```

```python
converted_time = convert_time_zone(today_datetime, from_timezone, to_timezone)
print(converted_time)
```
