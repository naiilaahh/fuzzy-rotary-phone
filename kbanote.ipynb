# Variable Arguments and Default Parameters

This notebook demonstrates how to handle an unknown number of arguments using the **splat operator** (`*args`) and how to define **default parameter values** in Python functions. Each code cell includes a short explanation and an example you can run directly.


```python
# splat operator for unknown number of arguments.
def sum(*args):
    result = 0
    for num in args:
        result += num

    return result

print(sum(1, 2, 3, 4))
```

```python
# use equal sign for default parameters
def multiplyBy(num1, multiplyBy=2):
    return num1 * multiplyBy

print(multiplyBy(4, 3))
```
