# Date/Time Overview

```python
import datetime
from datetime import date, datetime, time
```

```python
# Creating a time object
lastWeek = date(2026, 4, 22)

print(lastWeek)
```

```python
today = date.today()
print(today)
```

```python
# ability to subtract dates using datetime.
print(today - lastWeek)
```

```python
# datetime gives both date and time
right_now = datetime.now()
print(right_now)
```

```python
# with dates build with datetime module 

print(right_now.year)
print(right_now.month)
print(right_now.day)
print(right_now.hour)
print(right_now.minute)
print(right_now.second)
```

```python
time_right_now = time(21, 0, 0)
print(time_right_now)
```

```python
# strftime() - string format time - reformats time as a string
print(today.strftime('%A %B %d, %Y'))
```
